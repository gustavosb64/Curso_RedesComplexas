# Redes Complexas: correlação de grau


## Homofilia (homophily) ou assortative mixing   

### Correlação entre graus (Degree-degree correlation)


A assortatividade em redes é geralmente medida em termos de grau. Em algumas redes, nós de graus similares tendem a estar conectados.


#### Quantificando a assortatividade

* **Probabilidade condicional**: $P(k|k')$
    A estimativa desta distribuição condicional em redes do mundo real são muito difíceis devido ao número finito de redes e a presença de barulho.
    
    $$ P(k|k') = \frac{P(k,k')}{Pk'} $$
     
    * *__Problema__*: a maioria das redes seguema a distribuição da Lei de Potência. Essa lei tende a favorecer vértices com grau muito baixo: probabilidade alta de encontrar vértice de grau baixo. Populando uma rede grande, teremos um espaço muito vasto, o que dificulta o cálculo da probabilidade condicional da rede.
    
    
* **Grau médio dos vizinhos (knn de k)**: Uma alternativa é definir o grau médio dos vizinhos do nó $i$ como:
    
    $$k_{nn}(i)=\frac{1}{k_i}\sum_{j=1}^N A_{ij}k_j $$
    
    Na função do grau:
    
    $$k_{nn}(k) = \frac{1}{N_k}\sum_{i|k_i=k}k_{nn}(i) = \sum_{k'} k'P(k'|k) $$
    
    Equivale a uma esperança condicional.
    
    * Podemos classificar as redes de acordo com a atendência da curva $k$ vs $knn(k)$
    
    
    
    Embora a função knn(k) seja importante para a caracterização da assortatividade em redes, é frequentemente conveniente considerar apenas um número para a caputra da magniture das correlações.
    * _Possível medida_: **coeficiente de correlação de Pearson** dos graus ao final de uma aresta.
        
    * **Coeficiente de correlação de Pearson**:
        * Varia entre -1 e 1:
            * Próximo de 1: vértices de graus conectados estão conectados (**assortativa**
            * Próximo de -1: hubs tendem a se conectar a vértices não muito conectados
            * Próximo de 0: não tem correlação de grau (**dissortativa**)
            
        $$ \rho = \frac{cov(X,Y)}{\sigma_x\sigma_y}  $$
        
        A covariância pode ser definida como:
        
        $$ cov(X_i,X_j) = E[(X_i - E[X_j])(X_j - E[X_i])] = E[(X_i - \mu)(X_j - \mu)]$$
    
        Considerando a estrutura da rede:
    
        $$ cov(X_i,X_j) = \sum_i \sum_j \frac{A_{ij}(X_i - \mu)(X_j - \mu}{2M} $$
        
        A divisão por $2M$ normaliza pelo número de conexões. (O que é $M$ ??)
            
            $$ 2M = \sum k_i$$
            $$ 1 = \frac{\sum k_i}{2M} $$
            $$ \mu = \frac{\sum k_i X_i}{2M}$$
            
        O coeficiente de assortatividade considerando o grau:
        
            $$ \rho = \frac{\sum_{i=1}^N\sum_{j=1}^N ( A_{ij} - \frac{k_i k_j}{2M}) X_i X_j}{\sum_{i=1}^N\sum_{j=1}^N ( k_i \delta(i,j) - \frac{k_i k_j}{2M}) X_i X_j} $$
            
            $$ \downarrow \quad X_i = k_i $$
            
            $$ \rho = \frac{\sum_{i=1}^N\sum_{j=1}^N ( A_{ij} - \frac{k_i k_j}{2M}) k_i k_j}{\sum_{i=1}^N\sum_{j=1}^N ( k_i \delta(i,j) - \frac{k_i k_j}{2M}) k_i k_j} \quad -1 \leq \rho \leq 1$$
            
        onde $\delta(i,j)$ é o delta de Kronecker, ou seja, é 1 apenas onde $i=j$.
        
        Vale lembrar que o $X_i, X_j$ podem ser quaisquer valores associados ao vértice, não precisa ser apenas o grau.

#### Curiosidades

* Em redes sociais, há uma tendência da assortativdade ser _positiva_;
* Em redes tecnológicas, há tendência da assortatividade ser _baixa_: os hubs se conectam a hubs poucos conectados. Exemplo: Internet, sistemas de aeroportos.
* Em redes biológicas, a assortatividade também costuma ser _baixa_. Isto está relacionado à _robustez_.
